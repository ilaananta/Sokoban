```python
class Graph():
  
  def __init__(self):
    self.graph={}
  
  def vertex(self,v):
    if v in self.graph.keys():
       return False
    else:
      self.graph[v]=[]
      return True
  
  def edge(self,v1,v2):
    if v1 not in self.graph or v2 not in self.graph:
      return False
    else:
      self.graph[v1].append(v2)
      return True 
  
  def dfs(self, s, d, path, limit = None):
        path.append(s)
       
        if s == d:
            return True
        for vertice in self.graph[s]:
            
            if vertice not in path:
                x = None
                if limit: 
                    x = limit - 1 
                    if x<1:
                        return False
                if self.dfs(vertice, d, path,x ):
                  return True
                else:
                  continue    
        path.pop()
        return False
  
  def bfs(self,s,d):
     q=[[s]]
     while(q):
       
       l=q.pop(0)
       
       n=l[-1]
       
       if(n==d):
         return l
       child=self.graph[n]
      
       for c in child:
        temp=list(l)
        temp.append(c) 
        q.append(temp)
```

```python
def getpos(tu,dic):
    matrix=dic[tu]
    a=np.array(matrix)
    posBox=tuple(tuple(x) for x in np.argwhere((a=='*') | (a=='$')))
    posgoal=tuple(tuple(x) for x in np.argwhere((a=='*') | (a=='.') |(a=='+')))
    return posBox,posgoal
```

```python
def goal_pull_distace(my_state,dict,posGoal):
        matrix=dict[my_state]
        
        distance_to_goal=[[[float("inf")]*26 for i in range(26)]for i in range(len(posGoal))]
        goals=[]
       
        for i,goal in enumerate(posGoal):
            distance_to_goal[i][goal[1]][goal[0]]=0
            queue=[goal]
            while queue:
                current=queue.pop(0)
                possible_moves = [(1,0),(0,1),(0,-1),(-1,0)]
                for direction in possible_moves:
                 new_pos=(current[0]+direction[0],current[1]+direction[1])
                 player_pos=(current[0]+2*direction[0],current[1]+2*direction[1])
                 
                 if distance_to_goal[i][new_pos[1]][new_pos[0]]==float("inf"):
                    if matrix[new_pos[0]][new_pos[1]] not in['#'] and matrix[player_pos[0]][player_pos[1]] not in ['#']:
                        
                        distance_to_goal[i][new_pos[1]][new_pos[0]]=distance_to_goal[i][current[1]][current[0]]+1
                        queue.append(new_pos)
            goals.append(goal)
           
        return distance_to_goal,goals
```

```python
#https://baldur.iti.kit.edu/theses/SokobanPortfolio.pdf
#https://github.com/KnightofLuna/sokoban-solver
import math
from scipy.optimize import linear_sum_assignment
def getHeuristics(state,dic,posBox,posGoal,choice):
    if(choice=='m'):
        distance = 0
     #print(posBox,posGoal)
        completes = set(posGoal) & set(posBox)
        sortposBox = list(set(posBox).difference(completes))
        sortposGoals = list(set(posGoal).difference(completes))
        for i in range(len(sortposBox)):
            distance += (abs(sortposBox[i][0] - sortposGoals[i][0])) + (abs(sortposBox[i][1] - sortposGoals[i][1]))
    #print(distance,posWorker)
    #for i in range(len(sortposBox)):
    #    distance += (abs(sortposBox[i][0] - posWorker[0])) + (abs(sortposBox[i][1] - posWorker[1]))
    #print(distance)    
        return distance
    elif(choice=='h'):
        distance_to_goal,goals=goal_pull_distace(state,dic,posGoal)
        box_to_goal=np.zeros((len(posBox),len(posGoal)))
        for i,box in enumerate(posBox):
            for j,goal in enumerate(posGoal):
                box_to_goal[i][j]=distance_to_goal[j][box[1]][box[0]]
                
        box_to_goal[box_to_goal == np.inf] = 100000000000000000000
        row_ind,col_ind=linear_sum_assignment(box_to_goal)
        #Sum the values of the assignment
        return sum([box_to_goal[row_ind[i]][col_ind[i]] for i in range(len(row_ind))])
```
